```python
# 4.Print the of a number (e.g., 374 → 3+7+4) using only loops.
num = 374
summ = 0
for i in range(3):
    quo = num % 10
    print(quo, end = '+')
    num = num // 10
    summ = summ + quo
print(' = ', summ)
```
