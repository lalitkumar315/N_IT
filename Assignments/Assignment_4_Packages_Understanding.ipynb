1. random
    - randint
    - random
    - choice
    - sample
    - seed 

```python
import random
ls = (5,6,7,8,9)
```

```python
random.randint(1,6)
```

```python
random.random()
```

```python
random.choice(ls)
```

```python
random.sample(ls,3)
```

```python
random.seed(2)
random.sample(ls,4)
```

```python
random.seed(2)
random.sample(ls,4) # same random output as the seed is set to 2.
```

```python
random.seed(1)
random.sample(ls,4) # changed output bcz of seed value change.
```

```python
random.seed(1)
random.sample(ls,5) # sample limit changed to 5 but first 4 digits are same.
```
