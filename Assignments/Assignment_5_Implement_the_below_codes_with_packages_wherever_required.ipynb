```python
# wap ask the user take two number print subtraction multiplication and division
```

```python
first_number = eval(input('Enter the First Number : '))
second_number = eval(input('Enter the Second Number : '))

print(f'Subtraction of given two number is : {first_number-second_number}')
print(f'Multiplication of given two number is : {first_number*second_number}')
print(f'Divsion of given two number is : {first_number/second_number}')
```

```python
# wap ask the user take the radius find the area of circle
```

```python
import math

radius_of_circle = eval(input('Enter the radius of the circle : '))

print(f'Area of the Circle is : {round(math.pi * (radius_of_circle ** 2), 2)}')
```
