1. Ramesh’s basic salary is input through the keyboard. His dearness allowance is 40% of basic salary, and
 house rent allowance is 20% of basic salary. Write a program to calculate his gross salary.

```python
basic_salary = eval(input('Enter your basic salary: '))
dearness_allowance = basic_salary * 0.4     # 40%
house_rent_allowance = basic_salary * 0.2   # 20%
gross_salary = basic_salary + dearness_allowance + house_rent_allowance
print('Your gross salary is: ', gross_salary)
```

2. The distance between two cities (in km.) is input through the keyboard. Write a program to convert and
print this distance in meters, feet, inches and centimeters.

```python
distance_in_km = eval(input('Enter the distance between two cities in km : '))

distance_in_m = distance_in_km * 1000
distance_in_cm = distance_in_m * 100
distance_in_ft = distance_in_m * 3.28084
distance_in_inch = distance_in_ft * 12

print(f'Distance in meters is {distance_in_m} m')
print(f'Distance in centimeters is {distance_in_cm} cm')
print(f'Distance in feet is {distance_in_ft} ft')
print(f'Distance in inches is {distance_in_inch} inches')
```

3. If the marks obtained by a student in five different subjects are input through the keyboard, find out the
aggregate marks and percentage marks obtained by the student. Assume that the maximum marks that
can be obtained by a student in each subject is 100.

```python
subject_1 = eval(input('Enter marks for subject Math : '))
subject_2 = eval(input('Enter marks for subject Physics : '))
subject_3 = eval(input('Enter marks for subject Chemistry : '))
subject_4 = eval(input('Enter marks for subject Computer Science : '))
subject_5 = eval(input('Enter marks for subject English : '))
aggregate_mark = subject_1 + subject_2 + subject_3 + subject_4 + subject_5

print('Aggregate Mark : ', aggregate_mark)
print('Percentage : ', (aggregate_mark / 500) * 100)
```

4. Temperature of a city in Fahrenheit degrees is input through the keyboard. Write a program to convert this
temperature into Centigrade degrees

```python
fahrenheit = eval(input('Enter temperature in Fahrenheit : '))
print(f'Temperature in Centigrade is : {(fahrenheit - 32) * 5/9}')  #(32°F − 32) × 5/9 = 0°C
```

5. The length & breadth of a rectangle and radius of a circle are input through the keyboard. Write a program
to calculate the area & perimeter of the rectangle, and the area & circumference of the circle.

```python
length = eval(input('Enter the length of the rectangle : '))
breadth = eval(input('Enter the breadth of the rectangle : '))
radius = eval(input('Enter the radius of the circle : '))

print(f'Area of Rectangle is : {length * breadth}')
print(f'Perimeter of Rectangle is : {2 * (length + breadth)}')

print(f'Area of Circle is : {3.14 * (radius ** 2)}')
print(f'Circumference of Circle is : {2 * 3.14 * radius}')
```

6. Two numbers are input through the keyboard into two locations C and D. Write a program to interchange
the contents of C and D.

```python
C = eval(input('Enter the first number C : '))
D = eval(input('Enter the second number D : '))

tmp = C
C = D
D = tmp     # or without tmp it should be like C, D = D, C

print(f'After interchange: C = {C}, D = {D}')
```

7. If a five-digit number is input through the keyboard, write a program to reverse the number.

```python
number_str = input('Enter a five digit number : ')

reversed_number_str = number_str[::-1]      # this will reverse the string.
reversed_number = int(reversed_number_str)  # convert string to integer.

print(f'The reversed number is : {reversed_number}')
```

8. If a four-digit number is input through the keyboard, write a program to obtain the sum of the first and last

9. In a town, the percentage of men is 52. The percentage of total literacy is 48. If total percentage of literate
men is 35 of the total population, write a program to find the total number of illiterate men and women if
the population of the town is 80,000.

```python
total_population = 80000

men = total_population * .52                # total men is 52%
women = total_population - men

total_literate = total_population * .48     # total literacy is 48%

literate_men = total_population * .35       # literate men is 35%
literate_women = total_literate - literate_men

print(f'Number of Illiterate Men : {men - literate_men}')
print(f'Number of Illiterate Women : {women - literate_women}')
```

10. A cashier has currency notes of denominations 10, 50 and 100. If the amount to be withdrawn is input
through the keyboard in hundreds, find the total number of currency notes of each denomination the
cashier will have to give to the withdrawer.

```python
total_amount = eval(input('Enter the amount to be withdrawn in hundreds : '))

notes_100 = total_amount // 100             # qty divisible by 100
remaining_amount = total_amount % 100

notes_50 = remaining_amount // 50           # qty divisible by 50 from remaining
remaining_amount = remaining_amount % 50

notes_10 = remaining_amount // 10           # qty divisible by 10 from remaining

print(f'Number of 100 rupee notes: {notes_100}')
print(f'Number of 50 rupee notes: {notes_50}')
print(f'Number of 10 rupee notes: {notes_10}')
```
