```python
# wap take three number and find the avg.
n1 = 10
n2 = 20
n3 = 30
avg = (n1+n2+n3)/3
print('the average of {}, {} and {} is {}'.format(n1,n2,n3,avg))
print(f'the average of {n1}, {n2} and {n3} is {avg}')
```

```python
# wap ask a user calculate area of triangle 
pi = 3.14
radius = 20
area = pi * radius * radius
print('The area of triangle is {}, where pi is {} and radius is {}.'.format(area,pi,radius))
print(f'The area of triangle is {area}, where PI is {pi} and radius is {radius}.')
```

```python
# wap calculate the area of right angle triagle
breadth = 5
height = 6
area = 0.5 * breadth * height
print('The area of right angle triangle is {}, where breadth is {} and height is {}.'.format(area,breadth,height))
print(f'The area of right angle triangle is {area}, where breadth is {breadth} and height is {height}.')
```

```python
# wap calculate area of rectangle
length = 7
breadth = 8
area = length * breadth 
print('The area of rectangle is {}, where length is {} and breadth is {}.'.format(area,length,breadth))
print(f'The area of rectangle is {area}, where length is {length} and breadth is {breadth}.')
```

```python
# wap to ask the user take two variable
bill = 5000
tip = 100
total = bill + tip
print('The bill is {} and tip is {}, Total bill is : {}'.format(bill,tip,total))
print(f'The bill is {bill} and tip is {tip}, Total bill is : {total}')
```

```python
# wap to calculate tax
salary = 50000
tax_percent= 5
tax = salary * tax_percent/100
total_salary = salary - tax
print('Salary is {}, where tax percentage is {}%, So in-hand salary is {}'.format(salary,tax_percent,total_salary))
print(f'Salary is {salary}, where tax percentage is {tax_percent}%, So in-hand salary is {total_salary}')
```
