#Assignment-12: While loop

```python
# 1. Find the square of first five numbers
i = 1
while i <= 5:
    print(f'the square of {i} is {i*i}')
    i += 1
```

```python
# 2. Find the evens and odds of 5 random numbers
import random
i = 1
while i <= 5:
    num = random.randint(1,100)
    if num % 2 == 0:
        print(f'{num} is even')
    else:
        print(f'{num} is odd')
    i += 1
```

```python
# 3. Find the number of divisors of 75
i = 1
count = 0
while i <= 75:
    if 75 % i == 0:
        print(i, 'is a divisor of 75')
        count += 1
    i += 1
print(f'the number of divisors of 75 are {count}')
```

```python
# 4. Print the 7th table
i = 1
while i <= 10:
    print(f'7 x {i} = {7*i}')
    i += 1
```

```python
# 5. Find the sum of first 10 natural numbers
i = 1
sum = 0
while i <= 10:
    sum += i
    i += 1
print(f'the sum of first 10 natural numbers is {sum}')
```
