```python
str1 = 'ola ola ola'
count = 0
for i in range(len(str1)):
    print(str1[i])
    if str1[i] == 'a':
        count += 1
print('\n',count)
```

```python
str1 = 'ola ola ola'
sum = 0
# sum of indexes of 'a'
for i in range(len(str1)):
    if str1[i] == 'a':
        sum += i
        print(i)
print(sum)
```

```python
count = 0
o = False
l = False
str1 = 'ola ola ola'
for i in str1:
    if i == 'o':
        o = True
    if o == True and i == 'l':
        l = True
    if l == True and i == 'a':
        count += 1
        o = False
        l = False
print(count)
```
