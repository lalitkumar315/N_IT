# while Loop

- while loop we need 3 things separate
- initial point
- increament/Decrement
- stopping point


```
# wap to use while loop for printing 1 to 9
i = 1
while i <= 9:
    print(i)
    i = i+1
```

```
# test while loop
i = 1
while True:
    print(i)
    i = i+1
    if i == 10:
        break
```

```
# range(1,11,1)
i = 1
while i <= 10:
    print(i)
    i = i+1
```

```
# range(10,0,-1)
i = 10
while i >= 1:
    print(i)
    i = i-1
```

```
# range(-10,0,1)
i = -10
while i <= -1:
    print(i)
    i = i+1
```

```
# range(-1,-11,-1)
i = -1
while i >= -10:
    print(i)
    i = i-1
```

```
import random
i = 1
while True:
    x = random.randint(1,10)
    if x % 2 == 0:
        print(x, ' is even')
    else:
        print(x, ' is odd')
    i = i+1
    if i == 6:
        break
```
