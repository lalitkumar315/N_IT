```python
# Q1: Sum of Three Numbers
# Write a function `sum_three(a, b, c)` that returns the sum of the three numbers.

def sum(a,b,c):
    return a+b+c

n1=int(input('Enter first number: '))
n2=int(input('Enter second number: '))
n3=int(input('Enter third number: '))

print(f'sum of three number is {sum(n1,n2,n3)}')
```

```python
# Q2: Area of a Circle
# Write a function `area_circle(radius)` that returns the area of a circle. Use π = 3.14159

def area_circle(radius):
    return 3.14159*radius*radius

r = int(input('Enter radius of circle: '))

print(f'Area of circle is {area_circle(r)}')
```

```python
# Q3: Check Even or Odd
# Write a function `check_even_odd(num)` that prints whether a number is Even or Odd.

def check_even_odd(num):
    if num%2==0:
        print(f'{num} is even')
    else:
        print(f'{num} is odd')

for i in range(2):
    n = int(input('Enter number: '))
    check_even_odd(n)
```

```python
# Q4: Calculate BMI
# Write a function `calculate_bmi(weight, height)` that returns BMI value using the formula:
# BMI = weight / (height ** 2)

def calculate_bmi(weight, height):
    return weight/(height**2)

w = eval(input('Enter weight: '))
h = eval(input('Enter height: '))

print(f'BMI is {round(calculate_bmi(w,h),2)}')
```

```python
# Q5: Greet with Name and Age
# Write a function `greet(name, age)` that prints:
# Hello <name>, you are <age> years old

def greet(name, age):
    print(f'Hello {name}, you are {age} years old')

n = input('Enter name: ')
a = int(input('Enter age: '))
greet(n,a)
```

```python
# Q6: Find Maximum of Two Numbers
# Write a function `find_max(a, b)` that returns the greater of the two numbers.

def find_max(a, b):
    if a > b:
        print(f'{a} is greater.')
    else:
        print(f'{b} is greater.')

n1 = int(input('Enter first number: '))
n2 = int(input('Enter second number: '))
find_max(n1,n2)
```

```python
# Q7: Power Function
# Write a function `power(base, exponent)` that returns base raised to the power exponent.

def power(base, exponent):
    return base ** exponent

b = int(input('Enter base: '))
e = int(input('Enter exponent: '))

print(f'{b} to the power {e} is {power(b,e)}')
```

```python
# Q8: Calculate Simple Interest
# Write a function `simple_interest(principal, rate, time)` that returns the simple interest.
# Formula: (P × R × T) / 100

def simple_interest(principal, rate, time):
    return (principal*rate*time)/100

p = int(input('Enter principal: '))
r = int(input('Enter rate: '))
t = int(input('Enter time: '))

print(f'Simple interest is {simple_interest(p,r,t)}')
```

```python
# Q9: Check Leap Year
# Write a function `is_leap_year(year)` that returns True if the year is a leap year, otherwise
# False.

def is_leap_year(year):
    if year%4==0:
        return True
    else:
        return False

for i in range(2):
    y = int(input('Enter year: '))
    print(f'{y} is leap year: {is_leap_year(y)}')
```
