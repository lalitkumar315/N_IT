```python
**Postional-Only  vs  Keyword_Only**

- Postional only means the postion is fixed

    - It will not Expect the argument name

    - Before slash arguments called as **Postional only** arguments

    - So dont give argument name before slash

- Keyword only means the arguments Expects to be provide

        - After star all arguments are **keyword only arguments**

        - We should definitely provide the argument name

- Before slash : Postional only NO

- After star : keyword only   YES
```

**args vs kwargs**

- args: arguments

    - it is used to provide multiple inputs

    - with out use the argument name

    - the result will come in **tuple format**

- kwargs: keyword arguments

    - it is used to provide multiple inputs

    - with useing the argument name

    - the result will come in **dictionary format**

```python
def fun1(a,b,c):
    print(a,b,c)

fun1(1,2,3)
fun1(a=1,b=2,c=3)
```

```python
def fun2(*args):
    print(args)
fun2(1)  # 1 dimensional
fun2(1,2) # 2D
fun2(1,2,3,4)
```
