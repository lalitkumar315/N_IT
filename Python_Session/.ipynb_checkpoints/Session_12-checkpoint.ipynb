# For Loops

- multiple time use the same piece of code ia called as loop
- whenever we heared a term reuse then think about : Functin
- In Same way if we heared **multiple times** or **multiple iteration** then think about **loops**
- Two methods in loops
  - **for loop**
    - if we know in advance when to stop the loop then prefer for loop
    - here we know the condition to stop
  - **While loop**
    - if we don't know in advance when to stop the loop then prefer While loop
    - here we don't know the condition to stop

- any loop we need 3 things
    - Initial point
    - Increment or decrement
    - Condition to stop

**Case - 2: range(start,stop)**
- Start value same as start only
- by default increment by one
- last value is stop-1

```python
range(5) # range Stop at 5-1
range(2,5) # range start from 2 and stop at 5-1
range(50,53) # 51 52 53
```

```python
import random
for i in range(2,5):
    random.randint(1,100)
    print(f'the square of {i} is ',i*i)
```

**Case - 3: range(start, stop, step)**
- start is start value only
- but stop depends on step here
- if step is a positive value
    - then increment by step size only
    - last value = stop-1 (same as it is case2)
- if step is a negative value
    - then increment by step size only
    - last value = stop+1 (same as it is case2)

```python
for i in range(2,20,2):
    print(i)
```
