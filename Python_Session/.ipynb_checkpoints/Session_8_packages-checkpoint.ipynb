**Module**
- module is a single python file it contains,
    - Methods 
    - Functions 
    - Variables
    - Classes

**package**
- package is a collection of Modules
- collection of python files

**Analogy**
- Assume that module is one gift
- Package is a bag of gifts

**Methods vs Functions**
- Method and functions both are defined for some task with some name
- But Method define under one class
- Function is defined outside the class


**Two Errors**
- Module Not found error
  - means it is not installed n our laptop
- Attribute error
  - we know that module has functions methods and variables
  - that are not available under this module
- Whenever we see attribute error perform ditr on module name
