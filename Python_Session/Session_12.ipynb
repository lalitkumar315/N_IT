**Case - 2: range(start,stop)**
- Start value same as start only
- by default increment by one
- last value is stop-1

```python
range(5) # range Stop at 5-1
range(2,5) # range start from 2 and stop at 5-1
range(50,53) # 51 52 53
```

```python
import random
for i in range(2,5):
    random.randint(1,100)
    print(f'the square of {i} is ',i*i)
```

**Case - 3: range(start, stop, step)**
- start is start value only
- but stop depends on step here
- if step is a positive value
    - then increment by step size only
    - last value = stop-1 (same as it is case2)
- if step is a negative value
    - then increment by step size only
    - last value = stop+1 (same as it is case2)

```python
import random
def game():
    num = int(input('Enter a number'))
    s_num = random.randint(1,10)
    if s_num == num:
        print(f'Entered Number is : {num} . You Won')
        return
    else:
        print(f'Entered Number is : {num} . You Fail')

for i in range(3):
    game()
```

```python
# wap to get random for 5 numbers and find the sum of evens
import random
s = 0
for i in range(5):
    num = random.randint(1,100)
    if num % 2 == 0:
        s = s + num
        print(f'Number {num} is even.')
print(s)
```

```python
# for evens and odds
import random
s_e = 0
s_o = 0
for i in range(5):
    num = random.randint(1,100)
    if num % 2 == 0:
        s_e = s_e + num
        print(f'Number {num} is even.')
    else:
        s_o = s_o + num
        print(f'Number {num} is odd.')
print(f'Sum of Evens : {s_e}')
print(f'Sum of Odds : {s_o}')
```
